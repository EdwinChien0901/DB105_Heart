#合併標籤文檔

```python
import csv

with open('00.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    data = list(reader)

with open('01.csv', 'r') as file:
    reader = csv.reader(file)
    data1 = list(reader)
    
with open('02.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    data2 = list(reader)
```

```python
import pandas as pd

df = pd.DataFrame(columns=['con', 'tag'])

for i in range(len(data)):
    df.loc[i] = [data[i][0], data[i][1]]
```

```python
for i in range(len(data1)):
    df.loc[len(data)+i] = [data1[i][0], data1[i][1]]
```

```python
for i in range(len(data2)):
    df.loc[len(data)+len(data1)+i] = [data2[i][1], data2[i][0]]
```

```python
for i in range(len(df)):
    df.loc[i]['tag'] = df.loc[i]['tag'].replace(']', '')
    df.loc[i]['tag'] = df.loc[i]['tag'].replace('[', '')
    df.loc[i]['tag'] = df.loc[i]['tag'].replace('\"', '')
    df.loc[i]['tag'] = df.loc[i]['tag'].replace('\'', '')
    df.loc[i]['tag'] = df.loc[i]['tag'].replace(' ', '')
    df.loc[i]['tag'] = df.loc[i]['tag'].split(',')
    print(len(df.loc[i]['tag']))
```

```python
df = df.drop([0])
df = df.reset_index(drop=True)
```

EX:銀河洞

解法:把新地點加入大字典

```python
location = [] #要去空值 #等一下

for i in range(len(conlist)):
    tmp = []
    for j in range(len(conlist[i])):
        for k in range(len(raw)):
            if conlist[i][j] == raw[k]:
                tmp.append(raw[k])
    location.append(tmp)
```

去空值: https://stackoverflow.com/questions/3845423/remove-empty-strings-from-a-list-of-strings

```python
conindex = []

for i in range(len(location)):
    for j in range(len(location[i])):
        location[i][j] = location[i][j].replace(' ', '')
    location[i] = list(filter(None, location[i]))
    location[i] = list(set(location[i]))
    conindex.append(i)
```

```python
location
```

```python
empty = []

for i in range(len(location)):
    if len(location[i])<1:
        empty.append(i)
```

```python
with open('emptytag.txt', 'w', encoding='utf-8') as file:
    for i in empty:
        file.write(str(i)+'. ')
        file.write(df.loc[i][0])
        file.write('\n')
```

```python
dfz = pd.DataFrame(columns=taglist, index = conindex)

for i in range(len(df)):
    print(i)
    for j in range(len(df.loc[i]['tag'])):
        #print(df.loc[i]['tag'])
        for k in range(len(taglist)):
            #print(j, k)
            if df.loc[i]['tag'][j] == taglist[k]:
                #print(j, k)
                #print(taglist[k])
                #print(i, df.loc[i]['tag'][j], taglist[k])
                #print(dfz.loc[1])
                dfz.loc[i][taglist[k]] = 1
            else:
                if dfz.loc[i][taglist[k]] == 1:
                    continue
                else:
                    dfz.loc[i][taglist[k]] = 0
            
```
